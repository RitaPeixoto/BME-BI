```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
```

```python
r = requests.get("https://www.numbeo.com/cost-of-living/country_price_rankings?itemId=105", cookies={"srb_1":"1_wl"})

soup = BeautifulSoup(r.text, "html.parser")

salaries = soup.find_all("tr", class_="tr_standard")
print(salaries)
```

```python
countries = []
salaries_final = []
for salary in salaries:
    r = salary.find_all("td", class_="")
    rr = re.findall('(<td style="vertical-align: top">)\n([a-zA-Z ()]+)(<\/td>)',str(r[1]))
    country = rr[0][1]
    rr2 = re.findall('([0-9,.]+)', str(r[-1]))
    s = rr2[0]
    s = re.sub(',', '', s, count=0, flags=0)
    s = float(s)
    countries.append(country)
    salaries_final.append(s)
```

```python
print(countries)
print(salaries_final)
```

```python
result = []
for i in range(len(countries)):
    result.append([countries[i], float(salaries_final[i])])
print(result)
```

```python
df = pd.DataFrame(result, columns=['Country', 'Average Salary'])
df
```

```python
df.to_csv('datasets\CountriesAverageSalary.csv', sep=";", index=False)
```
