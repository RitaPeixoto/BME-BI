This dataset includes titles that contain the field separator in their field. As pentaho was not dealing well with this case, I've created this script that solves does cases to allow easy extraction on pentaho. This cases were not discarded as it consisted of 162 entries

```python
import pandas as pd 
import re
```

```python
df = pd.read_csv('datasets\MoviesOnStreamingPlatforms.csv')
```

```python
df.head(200)
```

```python
for i, r in df.iterrows():
    start = r['#']
    ret = re.compile("\"")
    prev_index = 0
    if ret.search(str(start)) != None :
        index = start.find(",")
        df.loc[i, "#"] = start[0: index]   
        prev_index = index  
        index = start.find(",", prev_index + 1)
        df.loc[i, "ID"] = start[prev_index + 1: index] 


        prev_index = index  
        aux = start.find("\"", prev_index + 1)
        index = start.find("\"", aux + 1)
        df.loc[i, "Title"] = start[aux + 1: index] 
        prev_index = index + 1  
        index = start.find(",", prev_index + 1)
        df.loc[i, "Year"] = start[prev_index + 1: index] 
        prev_index = index  
        index = start.find(",", prev_index + 1)
        df.loc[i, "Age"] = start[prev_index + 1: index] 
        prev_index = index  
        index = start.find(",", prev_index + 1)
        df.loc[i, "Rotten Tomatoes"] = start[prev_index + 1: index] 
        prev_index = index  
        index = start.find(",", prev_index + 1)
        df.loc[i, "Netflix"] = start[prev_index + 1: index] 
        prev_index = index  
        index = start.find(",", prev_index + 1)
        df.loc[i, "Hulu"] = start[prev_index + 1: index] 
        prev_index = index  
        index = start.find(",", prev_index + 1)
        df.loc[i, "Prime Video"] = start[prev_index + 1: index]
        prev_index = index  
        index = start.find(",", prev_index + 1)
        df.loc[i, "Disney+"] = start[prev_index + 1: index]
        prev_index = index  
        df.loc[i, "Type;"] = start[prev_index + 1: ]
    
df.rename(columns = {'Type;':'Type'}, inplace = True)
    
```

```python
df.to_csv('datasets\MoviesOnStreamingPlatforms_1.csv', sep=";", index=False)
```
